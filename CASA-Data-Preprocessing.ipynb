Columnns with Missing data:
    1. Merchant State
    2. Zip
    3. Error?
    4. user_Apartment

Among 44 columns there are only 4 columns with missing values.  But, 'user_apartment' may not be needed so dropping the user_apartment column

```python
df=df.drop('user_Apartment', axis=1)
```

```python
#removing the $ sign from the Amount column
columns=['Yearly Income - Person','Per Capita Income - Zipcode','Credit Limit','Amount','Total Debt']
for col in columns:
    df[col]=df[col].replace(r'[$]', '', regex=True)
```

Amount is supposed to be a numerical column, but the currency symbol in it makes it as a object which is not suitable to train a ML model.  So removed the currency symbol from the column

```python
df.head()
```

Removed all the currency symbols

It is observed that for the Online transactions we have missing 'Merchant State' and 'Zip'.  So for online transactions 'Merchant State' will be 'online' and 'Zip' will be '0000'

```python
df['Merchant State']=df['Merchant State'].fillna('online')
```

```python
df['Zip']=df['Zip'].fillna(0000)
```

```python
df['Errors?']=df['Errors?'].fillna('No Error')
```

```python
df.to_csv('credit_debit_preprocessed_transactions.csv')
```
